```python
csv = glob.glob("BD//*csv*")
caminho = csv[0]
print(caminho)
```

```python
df = pd.read_csv(caminho)
print(df.head())
```

```python
df = df.set_index('Year')
print(df)
```

```python
x = df.index
y = df.sum(axis='columns')
fig, ax = plt.subplots(figsize=(10,4))
ax.plot(x,y, label='Série anual de precipitação')
ax.set_ylim(500,2400)
ax.set_xlim(min(x),max(x))
ax.set_ylabel('Precipitação (mm)')
ax.set_xlabel('Ano')
ax.legend(loc='upper right')
ax.set_title('Precipitação acumulada ao longo dos anos')
```

```python
fig.savefig('BD//serie_precipitacao.pdf')
```
