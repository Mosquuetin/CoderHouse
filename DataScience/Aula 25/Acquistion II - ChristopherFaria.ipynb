```python
import pandas as pd
import requests

# Step 1: Read JSON from web API
response = requests.get('https://jsonplaceholder.typicode.com/users')

# Step 2: Check if the request was successful
if response.status_code == 200:
    data = response.json()
else:
    print("Failed to get data:", response.status_code)

# Step 3: Pass it to a DataFrame
df = pd.DataFrame(data)
```

```python
# REMOVENDO APENAS A PALAVRA 'street' DA COLUNA 'address'

# Remove 'street' from each address
df['address'] = df['address'].apply(lambda address: {key: val for key, val in address.items() if key != 'street'}) # .APLPLY É UM METODO PARA APLICAR UMA FUNÇÃO A CADA ELEMENTO DE UMA COLUNA
# NESTE CASO, APLICAMOS UMA FUNÇÃO LAMBDA QUE REMOVE A PALAVRA 'street' DE CADA LINHA DA COLUNA 'address'
```

```python
# Reload the data from the API
response = requests.get('https://jsonplaceholder.typicode.com/users')
if response.status_code == 200:
    data = response.json()
else:
    print("Failed to get data:", response.status_code)

# Recreate the DataFrame
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
# SEPARANDO AS CHAVES E VALORES DA COLUNA 'address' EM COLUNAS SEPARADAS

df_address = pd.DataFrame(df['address'].to_list()) # .to_list() transforma a coluna 'address' em uma lista

df_address

df_address.columns  # mostra as colunas da nova tabela

df_address.columns = ['address_' + col for col in df_address.columns] # renomeia as colunas da nova tabela

df_address
```

```python
# MERGE AS TUAS TABELAS USANDO A COLUNA ADDRESS COMO CHAVE

df = pd.concat([df, df_address], axis=1) # axis=1 significa que o merge será feito pelas colunas
```

```python
df.head()
```

```python
# REALIZANDO A MESMA OPERAÇÃO PARA A COLUNA 'company'

df_company = pd.DataFrame(df['company'].to_list())

df_company.columns = ['company_' + col for col in df_company.columns]

df_company

df = pd.concat([df, df_company], axis=1)    

df.head()
```

```python
# DROP AS COLUMNS 'address' E 'company'

df.drop(['address', 'company'], axis=1, inplace=True) # axis=1 significa que o drop será feito pelas colunas

df.head()
```

```python
# REPETIR A AÇÃO PARA A COLUNA 'address_geo' e depois DROP 'address_geo'

df_geo = pd.DataFrame(df['address_geo'].to_list())  # transforma a coluna 'geo' em uma lista

df_geo.columns = ['address_geo' + col for col in df_geo.columns] # renomeia as colunas da nova tabela  

df_geo  

df = pd.concat([df, df_geo], axis=1) # concat das tabelas

# drop das colunas 'address_geo' 

df.drop(['address_geo'], axis=1, inplace=True) # axis=1 significa que o drop será feito pelas colunas

df.head()
```
